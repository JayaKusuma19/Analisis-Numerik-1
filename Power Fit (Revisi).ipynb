```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
print(f"Diketahui Power Fit dengan bentuk y topi = Ax^M \nM merupakan bilangan yang telah diketahui \nAkan dicari nilai koefisien dari A")
```

```python
data_x_str = input("Masukkan data x, dipisahkan oleh spasi : ")
x = np.array(list(map(float, data_x_str.split())))
```

```python
data_y_str = input("Masukkan data y, dipisahkan oleh spasi : ")
y = np.array(list(map(float, data_y_str.split())))
```

```python
print("Data x:", x)
print("Data y:", y)
```

```python
M = float(input("Masukkan nilai M: "))
```

```python
result1 = np.sum(x**M * y)
```

```python
result2 = np.sum(x**(2*M))
```

```python
A = result1 / result2
```

```python
print(f"Nilai Dari A adalah: {A}")
```

```python
xpangkatM = x**M
```

```python
ytopi = A * xpangkatM
```

```python
plt.figure (figsize=(8,6))
plt.scatter(x,y,color='blue', label='Data (x,y)')
plt.scatter(x,ytopi,color='red', label='Data (x,y topi)')
plt.plot(x,y, linestyle='-',color='blue')
plt.plot(x,ytopi, linestyle='-',color='red')
plt.xlabel('Nilai x')
plt.ylabel('Nilai y dan y topi')
plt.title('Grafik Data (x,y) dan (x,y topi)')
plt.legend()
plt.grid(True)
plt.show
```
